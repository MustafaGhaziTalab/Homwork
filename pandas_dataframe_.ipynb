```python
import pandas as pd
# list 1
students = ['Alan', 'Vivian', 'Alister', 'Jade']

# list 2
age = [24, 26, 32, 29]

# using zip to merge the two lists
list_of_tuples = list(zip(students, age))

# list(zip(students, age)) will return
# [(Alan, 24), (Vivian, 26),(Alister, 32),(Jade, 29)]

# Converting the lists of tuples into pandas Dataframe.
df = pd.DataFrame(list_of_tuples,
columns=['Students', 'Age'])
print(df)
```

```python

import pandas as pd

# Define the dictionary with proper quotation marks
data = {'Student': ['Peter', 'James', 'Ella', 'Charlotte'],
        'Age': [28, 24, 35, 27],
        'Major': ['Chemistry', 'Biology', 'Physics', 'Chemistry']}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(data)
print(df)
```

```python
# adding a new row
df.loc[len(df.index)] = ['Mike', 33, 'Physics']
print(df)
```

```python

import pandas as pd
from numpy.random import randint

# Define a dictionary with correct quotes
data = {
    'Student': ['Peter', 'James', 'Ella', 'Charlotte'],
    'Age': [28, 24, 35, 27],
    'Major': ['Chemistry', 'Biology', 'Physics', 'Chemistry']
}

# Creating a DataFrame from the dictionary
df = pd.DataFrame(data)

# Define the new row
new_row = {'Student': 'Mike', 'Age': 29, 'Major': 'Biology'}

# Append the new row to the DataFrame
df = df.append(new_row, ignore_index=True)

# Print the updated DataFrame
print(df)
```
